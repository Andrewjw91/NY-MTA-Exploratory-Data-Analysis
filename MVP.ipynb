## Top 10 Train Stations By Average Number of Exits Per Day


Using the MTA turnstile data, I was able to determine the top 10 train stations with the highest average exits per day between the dates of 4/27/2019 to 7/26/2019. The barplot above is a visual representation of the collected data.

My analysis/model can be adapted to return up-to-date results, as new MTA data is published. This allows your company (Lime, Lyft, Bird etc.) to make sure your scooter fleet is being deployed at the current top 10 stations with the highest average exit traffic. Placing a larger number of scooters against higher foot traffic will in turn increase the potential for a rider, and the potential for recirculation of the scooters to subsequent riders.
